# Extracting and Analyzing text using regular expressions

```python
import requests 
from bs4 import BeautifulSoup
import urllib3
```

```python
r = requests.get('https://www.packtpub.com/support/faq')
r
```

We are able to load it through requests only, but since it asks us to do in urllib we will do that only 

```python
http = urllib3.PoolManager()
rr = http.request('GET', 'https://www.packtpub.com/support/faq')
rr.status
```

```python
len(rr.data)
```

```python
rr.data[:1000]
```

```python
soup = BeautifulSoup(rr.data)
```

```python
soup
```

# Questions

```python
questions = []
for div in soup.find_all('div',attrs={'class':'tab'}):
    q = div.label.text
    questions.append(q)
    print(q)
```

# Answers

```python
answers = []
for tag in soup.find_all('div',attrs = {'class':'tab-content'}):
    a = tag.text.strip()
    answers.append(a)
    print(a)
```

```python
len(answers)
```

# Create a df consisting of Questions and answers

```python
import pandas as pd 

df = pd.DataFrame({'Questions':questions,'Answers':answers})
df
```
