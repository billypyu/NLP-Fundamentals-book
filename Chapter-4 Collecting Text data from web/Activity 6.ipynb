# Extracting information from an online HTML page

```python
import requests 
from bs4 import BeautifulSoup
```

```python
result = requests.get('https://en.wikipedia.org/wiki/Rabindranath_Tagore')
```

```python
result
```

```python
soup = BeautifulSoup(result.content)
```

```python
soup
```

# List of heading under work section

```python
print('List of Headings under work section - ')
print('')
for tag in soup.find_all('h3')[:7]: 
    if tag.find('span',attrs = {'class':'mw-headline'}):
        print(tag.get_text())
```

```python
for tag in soup.find_all('h3')[8:]: 
    if tag.find('span',attrs = {'class':'mw-headline'}):
        print(tag.get_text())
```

# List of universities named 

```python
len(soup.find_all('ul'))
```

```python
preceding_text = soup.find('span',attrs={'id':'List_of_universities_and_university_buildings_named_after_him'})
```

```python
ul = preceding_text.find_next()
for row in ul.find_all('a'):
    print(row.text)
```
